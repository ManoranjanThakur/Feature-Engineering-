## Importing essential modules and files

```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm
df_adv = pd.read_csv('/kaggle/input/advertising/Advertising.csv', index_col=0)

df_adv.head()
```

```python
X=df_adv[['TV', 'radio', 'newspaper']]
y=df_adv['sales']
```

## fit a OLS model with intercept on TV and radio

```python
X=sm.add_constant(X)
model=sm.OLS(y,X).fit()
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:,1:].corr()
```

### In the above dataset we can see that there was not enough multicollinearity between the independent Varialbles.










### Now we will see this multicollinearity concept in a dataset which contains multicollinearity between independent variables

```python
df_salary=pd.read_csv('/kaggle/input/salary-data-with-age-and-experience/Salary_Data.csv')
df_salary.head()
```

```python
X=df_salary[['YearsExperience' , 'Age']]
```

```python
y=df_salary['Salary']
```

```python
X=sm.add_constant(X)
model=sm.OLS(y,X).fit()
```

```python
model.summary()
```

```python
X.iloc[:,1:].corr()
```

## Here you can see that these features are having much collinearity with each other.
