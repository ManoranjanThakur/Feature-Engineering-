## Some More Examples

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(6)
school_height=stats.poisson.rvs(loc=18,mu=35,size=1500)
classA_height=stats.poisson.rvs(loc=18,mu=30,size=60)
```

```python
classA_height.mean()
```

```python
_,p_value=stats.ttest_1samp(a=classA_height,popmean=school_height.mean())
```

```python
school_height.mean()
```

```python
if p_value < 0.5:
    print("We are accepting")
else:
    print("We are rejecting")
```
